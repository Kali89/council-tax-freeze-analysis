## This is not a novel mechanism

Before anything else: the redistributive effect of freezing council tax bands at 1991 values has already been established in print, by better-resourced researchers than us. If you know this literature, you should see us say so before you have to point it out.

**Adam, Hodge, Phillips & Xu (2020), *Revaluation and reform: bringing council tax in England into the 21st century*, IFS Report R169** (funded by the Nuffield Foundation), found that:

- Council tax has become **more regressive since 1991** because property values rose fastest in areas where they were already highest. Their Figure 2.4: average property prices in November 2019 were over **six times** their January 1995 level in London, versus barely **three times** in the North East.
- This is explicitly a **North/South story**, not incidental regional noise.
- Most importantly for this repository: because **central government's funding allocation to local authorities still references relative 1991 property values**, the frozen valuation date doesn't just distort council tax bills directly — it distorts the funding settlement built on top of it. In their words:

  > "councils in the North East must now levy more tax on a property worth (say) £250,000 than councils in London, if both are to deliver the spending on services deemed necessary by central government."

  That sentence is doing two jobs here. First, it is a published statement of the mechanism this whole repository quantifies. Second — and this matters for Variant 3, below — it is a published statement that **the unfairness could in principle be corrected through the funding settlement alone**, without ever touching council tax bands. That is the strongest form of the objection Variant 3 exists to test, and it has a citation, not just our intuition.

**Breach (2024), Centre for Cities, "Towards fiscal devolution"** (in *Devolution Solution: How fixing English local government will improve economic growth*) independently models a revaluation with additional top and bottom bands (an "A+" band below £125,000 and "I"/"J" bands above £2.5m and £5m). Two things carry over directly into our method:

- They hit the **same open-band imputation problem** we solve in `config.py` — what value do you assume for a band with no upper (or lower) bound? Their footnote states plainly that "more conservative values have been used for the modelling for these bands." We cite this as precedent for choosing the conservative corner of the plausible range ourselves, not as a coincidence.
- They argue that **revaluation without fiscal devolution risks minimal or zero fairness gain for large political cost** — relevant context for why the Variant 3 answer matters practically, not just statistically.

### What this repository adds

Both of the above are **snapshot counterfactuals**: what happens if bands are revalued *now*, with distributional analysis of winners and losers on revaluation day. Neither publishes the **cumulative magnitude of the freeze's redistributive effect as a local-authority-level time series**.

"London would pay £X more after revaluation" is a different claim from "London has underpaid £Y per dwelling cumulatively since 2000." The first is a flow, evaluated at a point in time. The second is the accumulated stock of that flow, integrated over a quarter-century of divergent house price growth. **We are quantifying the stock, not re-discovering the flow** — and Section 4 (Variant 3) uses the flow-level IFS finding as an external check on whether our stock estimate is even the right order of magnitude, before we trust it enough to publish it.

## The study period: a headline series and a labelled backward extension

The counterfactual needs `count[i,b,t]`, dwelling counts by band, at the same geographic resolution as the Band D rate it's paired with. For most of 2000-01 to 2025-26 that resolution matches cleanly. It doesn't for one specific, bounded case, and rather than quietly patch around it, the mismatch is handled by splitting the reported series in two.

**The problem.** Seven local authorities were created by the 2009 reorganisation - Cornwall, County Durham, Northumberland, Shropshire, Wiltshire, Bedfordshire (as two authorities), Cheshire (as two authorities) - from 37 predecessor districts. MHCLG's Band D data preserves each predecessor's own rate for the years before 2009 (six separate rates for pre-2009 Durham, for instance). VOA's CTSOP dwelling-band data does not: it retroactively aggregates the 2009-wave predecessors into their successor's count, with no predecessor-level band breakdown recoverable from that source. Combining predecessor-level rates correctly requires predecessor-level weights, and for band-level detail, none exist in the published record.

**How big a problem.** The seven 2009-wave areas held ~6.3% of England's dwelling stock throughout 2000-2008 - cross-checked from two independent sources (VOA CTSOP's own retroactive totals, and MHCLG's separately-published Dwelling Stock Estimates), which agree to within a few tenths of a percentage point. A 6.3% area share alone might be a rounding footnote. It isn't, because of what's inside it: County Durham's six recoverable predecessor rates for 2000-01 span £867.69 to £1,009.34 - a ~16% range - meaning the weighting choice used to combine them materially moves County Durham's reconstructed liability. Durham and Northumberland are North East authorities, sitting directly on the low-appreciation end of the North/South comparison this analysis exists to measure. An error there does not average out against the South; it lands on the comparison itself.

**What was checked, not assumed, before deciding what to do about it.** Individual historic VOA releases from 2000-2009 with predecessor-district rows intact were searched for directly - the UK Government Web Archive's holdings for `voa.gov.uk` across that whole period, every plausible filename pattern - and not found; VOA's own statistics publishing infrastructure appears to date from around 2010. A fallback does exist, partially: MHCLG's Live Table 125 (Dwelling Stock Estimates by local authority) carries genuine predecessor-level **total** dwelling counts back to 2001 for all 37 predecessor districts - unbanded. Real weights, no band split.

**The resolution:**

- **Headline series: 2009-10 to 2025-26.** Every local authority on observed CTSOP band counts. Zero imputation. This is the load-bearing claim in this repository, and it is clean. Seventeen years is still comfortably the longest cumulative estimate of this mechanism published anywhere - both IFS (2020) and Centre for Cities are snapshots, not time series, at all (see above).
- **Backward extension: 2000-01 to 2008-09**, reported *separately*, never folded into the headline: Table 125's predecessor-level dwelling totals, combined with a *back-projected band split* - each county's earliest observed post-2009 CTSOP band-share percentages, applied to its predecessors' Table 125 totals. Reported as "extending backward on imputed band shares adds a further £X per dwelling."

**The imputation's bias is conservative - the same structural property as the Band A/H midpoint choice, and stated with the same directness.** Assigning Easington or Penwith their county's post-2009 average band mix gives them a *better* mix than they likely really had, since poorer predecessor districts pull the true county average down. A better band mix implies a higher imputed 1991-basis stock value, which implies a higher counterfactual liability, which implies a *smaller* measured gap. The backward extension is therefore a **floor** on the 2000-09 period's contribution, not a central estimate - exactly the same "conservative corner" framing used for the Band A/H midpoint ratios, not a separate argument invented for this case.

**Sensitivity, not a single imputed number.** The back-projected band shares are swept across three methods - county-average earliest-observed (base case), successor's-earliest-observed (distinguishes the two-successor counties, Bedfordshire and Cheshire, from the single-successor base case), and a deliberately pessimistic skew for low-value predecessors - and the extension's movement across methods is reported regardless of outcome. If it barely moves, that itself is a finding worth reporting, and a stronger one than a single unchallenged number would be.

**Why split it this way.** A hostile reader can reject the entire 2000-09 extension and the headline survives completely untouched - it just gets shorter. The contestable material (the imputation) sits outside the load-bearing claim, which is where contestable material belongs. An alternative considered and dropped: coarsening the dwelling-count weighting scheme for *all* 26 years to avoid a single imputation confined to 9 of them. That would worsen resolution everywhere to fix a problem that only exists for seven local authorities in nine years, and would introduce a pre/post-2009 methodological inconsistency in exchange - two problems for the price of one.

**A small piece of corroboration, noted for the record.** Table 125's Old/New ONS code columns - built for an entirely different purpose, MHCLG's own dwelling-stock accounting, never for boundary harmonisation - independently reproduce the Phase 1 crosswalk's GSS code resolutions exactly (Penwith: old code `15UF`, new code `E07000023`, matching the ONS Code History Database). Unplanned confirmation from a second, unrelated government series that the boundary harmonisation work is correct, not merely self-consistent.
