## This is not a novel mechanism

Before anything else: the redistributive effect of freezing council tax bands at 1991 values has already been established in print, by better-resourced researchers than us. If you know this literature, you should see us say so before you have to point it out.

**Adam, Hodge, Phillips & Xu (2020), *Revaluation and reform: bringing council tax in England into the 21st century*, IFS Report R169** (funded by the Nuffield Foundation), found that:

- Council tax has become **more regressive since 1991** because property values rose fastest in areas where they were already highest. Their Figure 2.4: average property prices in November 2019 were over **six times** their January 1995 level in London, versus barely **three times** in the North East.
- This is explicitly a **North/South story**, not incidental regional noise.
- Most importantly for this repository: because **central government's funding allocation to local authorities still references relative 1991 property values**, the frozen valuation date doesn't just distort council tax bills directly — it distorts the funding settlement built on top of it. In their words:

  > "councils in the North East must now levy more tax on a property worth (say) £250,000 than councils in London, if both are to deliver the spending on services deemed necessary by central government."

  That sentence is doing two jobs here. First, it is a published statement of the mechanism this whole repository quantifies. Second — and this matters for Variant 3, below — it is a published statement that **the unfairness could in principle be corrected through the funding settlement alone**, without ever touching council tax bands. That is the strongest form of the objection Variant 3 exists to test, and it has a citation, not just our intuition.

**Breach (2024), Centre for Cities, "Towards fiscal devolution"** (in *Devolution Solution: How fixing English local government will improve economic growth*) independently models a revaluation with additional top and bottom bands (an "A+" band below £125,000 and "I"/"J" bands above £2.5m and £5m). Two things carry over directly into our method:

- They hit the **same open-band imputation problem** we solve in `config.py` — what value do you assume for a band with no upper (or lower) bound? Their footnote states plainly that "more conservative values have been used for the modelling for these bands." We cite this as precedent for choosing the conservative corner of the plausible range ourselves, not as a coincidence.
- They argue that **revaluation without fiscal devolution risks minimal or zero fairness gain for large political cost** — relevant context for why the Variant 3 answer matters practically, not just statistically.

### What this repository adds

Both of the above are **snapshot counterfactuals**: what happens if bands are revalued *now*, with distributional analysis of winners and losers on revaluation day. Neither publishes the **cumulative magnitude of the freeze's redistributive effect as a local-authority-level time series**.

"London would pay £X more after revaluation" is a different claim from "London has underpaid £Y per dwelling cumulatively since 2000." The first is a flow, evaluated at a point in time. The second is the accumulated stock of that flow, integrated over a quarter-century of divergent house price growth. **We are quantifying the stock, not re-discovering the flow** — and Section 4 (Variant 3) uses the flow-level IFS finding as an external check on whether our stock estimate is even the right order of magnitude, before we trust it enough to publish it.

## Variant 1 is the headline; Variant 2 is a separate finding, not a bigger version of it

Both variants share one calculation, not two: for every dwelling cohort, we compute a single relative value (how a 1991 band's assumed midpoint has moved, relative to the *national average*, since the baseline), and pass it through one of two liability-rate functions - a smooth curve that reproduces today's compressed 6/9-18/9 multiplier structure (Variant 1), or a straight proportional line with no compression at all (Variant 2, literally "Variant 1 with the compression removed"). Both then feed the same revenue-neutral reallocation. This matters for what follows: because they share a mechanism, we can say precisely what separates them, rather than comparing two different models that happen to point the same way.

**They are not interchangeable, and reporting them as one number would be wrong.** Checking where Variant 1 and Variant 2 disagree in *sign* - not a bug, see below - shows exactly what each one is picking up. The LAs where compression alone (the difference between the two) is largest are Elmbridge, Guildford, Mole Valley, Sevenoaks: Surrey commuter towns with 3-4x the national share of dwellings in Bands F-H. Their Variant 1 gap - the effect of the frozen valuation date alone - is small; their Variant 2 gap is large. That is a **rich-household story that cuts across regions**, not the North/South freeze story this analysis is about. Folding it into a single headline number would let a genuinely different mechanism (the 3:1 compression cap protecting expensive property everywhere, not just in the North's low-appreciation LAs) masquerade as the valuation-freeze finding.

**So: Variant 1 (frozen valuation date, existing multiplier structure) is the headline. Variant 2 (add full proportionality) is reported as a distinct, secondary finding** - what compression alone is worth, on top of the freeze, concentrated in different places for a different reason. The two are never summed or averaged into one figure.

**A note on what "disagreement" between the variants means, since it looks like a modelling failure until you check it.** Checked directly across all 296 LAs for FY2018-19: only 80% agree in sign on which way an LA's gap points. The mathematics behind the two liability-rate functions is clean - they cross exactly once, at Band D's own midpoint, verified numerically - so this isn't noise from a broken calculation. It's what "V2 = V1 + compression" predicts when the compression term is large enough to dominate a small valuation-date effect: the disagreement correlates at r = -0.69 with an LA's share of dwellings in Bands F-H across all 296 LAs, and the disagreeing LAs are exactly the Surrey-type commuter belt just described. A hostile reader who assumes Variant 1 and Variant 2 should always agree in sign will find this section before they find a discrepancy list - which is the point.

## Validating the engine against IFS (2020)

Before trusting a single cumulative number, the engine's implied FY2018-19 LA-level redistribution was checked against IFS (2020)'s own published figures for named local authorities - a gate, not a checkpoint: the instruction going in was to stop and report a failure, not tune the model until it passed. It passed. **Variant 1 vs IFS's own "pure revaluation" option: 14 of 15 named LAs agree on sign** (the one miss, Bristol, is IFS's own smallest listed effect - the kind of borderline case a band-level model is least equipped to get right). **Variant 2 vs IFS's "continuous and proportional" option: 8 of 8 agree on sign.**

**Sign agreement is not sufficient, and treating it as sufficient would have let a real, systematic error through.** Checking magnitudes, not just signs, on the same FY2018-19 data: Kensington and Chelsea and Westminster's Variant 2 gaps came out roughly 2-3x larger than IFS's own figures. Both are real LAs, not edge cases, and the gate had already passed on sign - which is exactly the failure mode a sign-only check cannot catch.

**What the investigation found.** Two things were checked and ruled out first: Westminster's Band D rate (£712 for 2018-19) is correct, verified against Westminster's own published council tax guide; and our implied average property value for Westminster (£1,253,118) is within 7% of IFS's own independent Q1-2019 estimate (£1,176,000). The stock-value calculation is not the problem. What explains it: **this engine reallocates each year's total actual revenue against value share in a single national pot - the brief's own specified formula.** The real system does not. Most of a London borough's bill (GLA precept: mayoral, police, fire) is identical across all 33 boroughs and was never value-proportional to begin with - it is reallocated, in reality, *within* London, not against the whole of England. Westminster's own (district-tier) rate is £712, against £933-1,707 for every other London borough sharing the same GLA/police/fire tier - a genuine, large, local anomaly (rooted in decades of business-rate income and reserves, not the valuation freeze), which the single-pot method takes and amplifies against the whole of England rather than just its own precepting peers. Checked directly: the ratio between an LA's stock-value share and its actual-revenue share - which single-pot reallocation uses directly - is 6.25x for Westminster, 4.10x for Kensington and Chelsea, 2.28x for Camden: a clean gradient tracking exactly how far each borough's own rate diverges from its shared-tier peers, not a London-wide effect.

**The fix - reallocating each precepting tier (district/county/police/fire/GLA) separately, within its own geography, the way IFS's own methodology does - needs data this pipeline does not yet have**, and was scoped rather than assumed: MHCLG does publish individual-authority precept-tier rates, with GSS codes, confirmed directly back to 2011-12 in a clean, structured format. 2009-10 and 2010-11 are not available in that form - the only likely source is an unstructured, multi-hundred-page departmental statistical digest, a materially different extraction task from everything else in this pipeline, not pursued.

**Decision: single-pot reallocation stays for the whole headline period, with the bias quantified per LA rather than corrected.** Tier-correcting only London while leaving the rest of England single-pot was considered and explicitly rejected: it would make the North/South comparison - the entire point of this analysis - a comparison between two different calculations, which is worse than a single method applied uniformly. A per-LA **exposure bound** is reported instead: `shared_tier_share = 1 - (own district precept / area total)`, computed directly from data already in the pipeline (no new source needed). It is structurally bimodal, not a matter of London vs the North: unitary and London authorities (whose own precept already absorbs what would be county-level services elsewhere) show 6-40%; ordinary two-tier shire districts (whose own precept is a small slice next to the county's) show 83-92%, in the North and South alike. Hartlepool, County Durham and Blackpool - all unitary authorities, all in the low-exposure band (14-16%) - are consistent with their own rates not diverging sharply from their (very small) set of shared-tier peers, which is why the Northern figures in this analysis are not subject to the same caveat as Westminster's. This bound does not prove a given LA's gap is unbiased, only how exposed it could be; it is reported alongside every gap, not used to silently adjust any of them.

**One piece of this is already done, not just argued for: the Price Paid calibration check.** Rather than pull the 0.75x/1.5x midpoint ratios from the air, 1995-97 Land Registry Price Paid sales in four local authorities were deflated to April-1991-equivalent values (a single *national* quarterly deflator - Nationwide's "UK house prices since 1952" series - not an LA-level one, keeping this a small, self-contained assumption rather than reopening the 1991-1995 bridging question ruled out for the main pipeline) and compared against the assumed thresholds:

| District | Band H empirical ratio (assumed 1.5) | n sales | Band A empirical ratio (assumed 0.75) | n sales |
|---|---|---|---|---|
| Kensington and Chelsea | **2.06** | 3,860 | n/a (0 sales below £40k) | 0 |
| City of Westminster | **1.78** | 2,287 | n/a (0 sales below £40k) | 0 |
| Blackpool | 2.66 (too few sales to trust) | 3 | **0.77** | 3,799 |
| Easington | 1.06 (too few sales to trust) | 1 | **0.64** | 2,092 |

At thousands of sales each, Kensington and Chelsea / Westminster's true Band H tail sits well above the assumed 1.5x - the ratio understates high-value stock. At thousands of sales each, Blackpool / Easington's true Band A tail sits at or below the assumed 0.75x, Easington notably so. Both findings push the same direction: the assumed ratios make the measured gap smaller than a more empirically-grounded pair would, not larger.
