## This is not a novel mechanism

Before anything else: the redistributive effect of freezing council tax bands at 1991 values has already been established in print, by better-resourced researchers than us. If you know this literature, you should see us say so before you have to point it out.

**Adam, Hodge, Phillips & Xu (2020), *Revaluation and reform: bringing council tax in England into the 21st century*, IFS Report R169** (funded by the Nuffield Foundation), found that:

- Council tax has become **more regressive since 1991** because property values rose fastest in areas where they were already highest. Their Figure 2.4: average property prices in November 2019 were over **six times** their January 1995 level in London, versus barely **three times** in the North East.
- This is explicitly a **North/South story**, not incidental regional noise.
- Most importantly for this repository: because **central government's funding allocation to local authorities still references relative 1991 property values**, the frozen valuation date doesn't just distort council tax bills directly — it distorts the funding settlement built on top of it. In their words:

  > "councils in the North East must now levy more tax on a property worth (say) £250,000 than councils in London, if both are to deliver the spending on services deemed necessary by central government."

  That sentence is doing two jobs here. First, it is a published statement of the mechanism this whole repository quantifies. Second — and this matters for Variant 3, below — it is a published statement that **the unfairness could in principle be corrected through the funding settlement alone**, without ever touching council tax bands. That is the strongest form of the objection Variant 3 exists to test, and it has a citation, not just our intuition.

**Breach (2024), Centre for Cities, "Towards fiscal devolution"** (in *Devolution Solution: How fixing English local government will improve economic growth*) independently models a revaluation with additional top and bottom bands (an "A+" band below £125,000 and "I"/"J" bands above £2.5m and £5m). Two things carry over directly into our method:

- They hit the **same open-band imputation problem** we solve in `config.py` — what value do you assume for a band with no upper (or lower) bound? Their footnote states plainly that "more conservative values have been used for the modelling for these bands." We cite this as precedent for treating the open bands as something to calibrate against real evidence (see the Price Paid check, below, and Sensitivity), not as a licence to guess.
- They argue that **revaluation without fiscal devolution risks minimal or zero fairness gain for large political cost** — relevant context for why the Variant 3 answer matters practically, not just statistically.

### What this repository adds

Both of the above are **snapshot counterfactuals**: what happens if bands are revalued *now*, with distributional analysis of winners and losers on revaluation day. Neither publishes the **cumulative magnitude of the freeze's redistributive effect as a local-authority-level time series**.

"London would pay £X more after revaluation" is a different claim from "London has underpaid £Y per dwelling cumulatively since 2000." The first is a flow, evaluated at a point in time. The second is the accumulated stock of that flow, integrated over a quarter-century of divergent house price growth. **We are quantifying the stock, not re-discovering the flow** — and Section 4 (Variant 3) uses the flow-level IFS finding as an external check on whether our stock estimate is even the right order of magnitude, before we trust it enough to publish it.

## Variant 1 is the headline; Variant 2 is a separate finding, not a bigger version of it

Both variants share one calculation, not two: for every dwelling cohort, we compute a single relative value (how a 1991 band's assumed midpoint has moved, relative to the *national average*, since the baseline), and pass it through one of two liability-rate functions - a smooth curve that reproduces today's compressed 6/9-18/9 multiplier structure (Variant 1), or a straight proportional line with no compression at all (Variant 2, literally "Variant 1 with the compression removed"). Both then feed the same revenue-neutral reallocation. This matters for what follows: because they share a mechanism, we can say precisely what separates them, rather than comparing two different models that happen to point the same way.

**They are not interchangeable, and reporting them as one number would be wrong.** Checking where Variant 1 and Variant 2 disagree in *sign* - not a bug, see below - shows exactly what each one is picking up. The LAs where compression alone (the difference between the two) is largest are Elmbridge, Guildford, Mole Valley, Sevenoaks: Surrey commuter towns with 3-4x the national share of dwellings in Bands F-H. Their Variant 1 gap - the effect of the frozen valuation date alone - is small; their Variant 2 gap is large. That is a **rich-household story that cuts across regions**, not the North/South freeze story this analysis is about. Folding it into a single headline number would let a genuinely different mechanism (the 3:1 compression cap protecting expensive property everywhere, not just in the North's low-appreciation LAs) masquerade as the valuation-freeze finding.

**So: Variant 1 (frozen valuation date, existing multiplier structure) is the headline. Variant 2 (add full proportionality) is reported as a distinct, secondary finding** - what compression alone is worth, on top of the freeze, concentrated in different places for a different reason. The two are never summed or averaged into one figure.

**A note on what "disagreement" between the variants means, since it looks like a modelling failure until you check it.** Checked directly across all 296 LAs for FY2018-19: only 80% agree in sign on which way an LA's gap points. The mathematics behind the two liability-rate functions is clean - they cross exactly once, at Band D's own midpoint, verified numerically - so this isn't noise from a broken calculation. It's what "V2 = V1 + compression" predicts when the compression term is large enough to dominate a small valuation-date effect: the disagreement correlates at r = -0.69 with an LA's share of dwellings in Bands F-H across all 296 LAs, and the disagreeing LAs are exactly the Surrey-type commuter belt just described. A hostile reader who assumes Variant 1 and Variant 2 should always agree in sign will find this section before they find a discrepancy list - which is the point.

## Validating the engine against IFS (2020)

Before trusting a single cumulative number, the engine's implied FY2018-19 LA-level redistribution was checked against IFS (2020)'s own published figures for named local authorities - a gate, not a checkpoint: the instruction going in was to stop and report a failure, not tune the model until it passed. It passed. **Variant 1 vs IFS's own "pure revaluation" option: 14 of 15 named LAs agree on sign** (the one miss, Bristol, is IFS's own smallest listed effect - the kind of borderline case a band-level model is least equipped to get right). **Variant 2 vs IFS's "continuous and proportional" option: 8 of 8 agree on sign.**

**Sign agreement is not sufficient, and treating it as sufficient would have let a real, systematic error through.** Checking magnitudes, not just signs, on the same FY2018-19 data: Kensington and Chelsea and Westminster's Variant 2 gaps came out roughly 2-3x larger than IFS's own figures. Both are real LAs, not edge cases, and the gate had already passed on sign - which is exactly the failure mode a sign-only check cannot catch.

**What the investigation found.** Two things were checked and ruled out first: Westminster's Band D rate (£712 for 2018-19) is correct, verified against Westminster's own published council tax guide; and our implied average property value for Westminster (£1,253,118) is within 7% of IFS's own independent Q1-2019 estimate (£1,176,000). The stock-value calculation is not the problem. What explains it: **this engine reallocates each year's total actual revenue against value share in a single national pot - the brief's own specified formula.** The real system does not. Most of a London borough's bill (GLA precept: mayoral, police, fire) is identical across all 33 boroughs and was never value-proportional to begin with - it is reallocated, in reality, *within* London, not against the whole of England. Westminster's own (district-tier) rate is £712, against £933-1,707 for every other London borough sharing the same GLA/police/fire tier - a genuine, large, local anomaly (rooted in decades of business-rate income and reserves, not the valuation freeze), which the single-pot method takes and amplifies against the whole of England rather than just its own precepting peers. Checked directly: the ratio between an LA's stock-value share and its actual-revenue share - which single-pot reallocation uses directly - is 6.25x for Westminster, 4.10x for Kensington and Chelsea, 2.28x for Camden: a clean gradient tracking exactly how far each borough's own rate diverges from its shared-tier peers, not a London-wide effect.

**The fix - reallocating each precepting tier (district/county/police/fire/GLA) separately, within its own geography, the way IFS's own methodology does - needs data this pipeline does not yet have**, and was scoped rather than assumed: MHCLG does publish individual-authority precept-tier rates, with GSS codes, confirmed directly back to 2011-12 in a clean, structured format. 2009-10 and 2010-11 are not available in that form - the only likely source is an unstructured, multi-hundred-page departmental statistical digest, a materially different extraction task from everything else in this pipeline, not pursued.

**Decision: single-pot reallocation stays for the whole headline period, with the bias quantified per LA rather than corrected.** Tier-correcting only London while leaving the rest of England single-pot was considered and explicitly rejected: it would make the North/South comparison - the entire point of this analysis - a comparison between two different calculations, which is worse than a single method applied uniformly. A per-LA **exposure bound** is reported instead: `shared_tier_share = 1 - (own district precept / area total)`, computed directly from data already in the pipeline (no new source needed). It is structurally bimodal, not a matter of London vs the North: unitary and London authorities (whose own precept already absorbs what would be county-level services elsewhere) show 6-40%; ordinary two-tier shire districts (whose own precept is a small slice next to the county's) show 83-92%, in the North and South alike. Hartlepool, County Durham and Blackpool - all unitary authorities, all in the low-exposure band (14-16%) - are consistent with their own rates not diverging sharply from their (very small) set of shared-tier peers, which is why the Northern figures in this analysis are not subject to the same caveat as Westminster's. This bound does not prove a given LA's gap is unbiased, only how exposed it could be; it is reported alongside every gap, not used to silently adjust any of them.

## Exposure alone isn't enough: which LAs are actually at risk, and how much does a NaN mean

`shared_tier_share` (above) is a bound, not a diagnosis: it tells you how much of a bill flows through tiers that single-pot reallocation could mishandle, not whether it actually does. That needs a second condition - does the LA's own (district/unitary) precept actually diverge, in cash terms, from the other LAs sharing its non-own tiers (`boundaries.precepting_groups.PRECEPTING_GROUP`: shire county, metropolitan county, or Greater London, sourced from ONS's real December-2024 LAD-to-county lookup)?

**The first version of this was wrong, and it's worth recording why rather than quietly shipping the fix.** Built as a ratio - own precept ÷ peer-group median own precept - it ranked *ordinary shire districts above Westminster*: Oxford scored higher than Westminster because Oxford's own precept (£303.80) is ~50% above its Oxfordshire peers' (£202.86 median) - a large *ratio* on a small base (£101 in cash terms, against a ~£1,900 total bill). Westminster's own precept (£417.86) is 64% *below* its Greater London peers' (a smaller ratio, 0.36), but on a much larger base - a **£755.97 cash gap**, against a bill of only £712.09. The reallocation this is meant to explain sums actual pounds and redistributes them by value share; the *distortion* it produces is therefore a pound quantity. A ratio answers "how proportionally unusual is this council's spending," which is a real fact about Oxford and irrelevant to whether single-pot mangles its gap. Replaced outright with a cash-share statistic, not kept alongside it: `single_pot_bias_risk = |own precept − peer-group median own precept| ÷ own area-total bill`.

**This independently reproduces the outlier set the original hand investigation found**, by a completely different route - two paths landing on the same five LAs is the reason it's trusted:

| LA | shared_tier_share | cash divergence from peers | single_pot_bias_risk |
|---|---|---|---|
| Westminster | 41.3% | −£755.97 | **1.06** |
| Wandsworth | 40.7% | −£745.41 | **1.03** |
| Hammersmith and Fulham | 28.8% | −£446.02 | 0.44 |
| City of London | 8.2% | −£316.52 | 0.34 |
| Kensington and Chelsea | 25.8% | −£328.65 | 0.29 |

Westminster and Wandsworth form a genuinely distinct top tier - both have a cash divergence from peers worth *more than their own entire discounted bill* (risk > 1.0), roughly 2.4x the next-highest LA - not a smooth decline down a single list. **Checked, not assumed: Wandsworth's computed gap is inflated the same way Westminster's is**, not just its risk score - both variants' percentage gaps sit within a few points of each other (Variant 1: −64.7% vs −63.8%; Variant 2: −84.0% vs −75.2%). The caveat below therefore names two severity tiers, not one flat list of five: **Westminster and Wandsworth as the severe pair, Hammersmith and Fulham / City of London / Kensington and Chelsea as a second, materially smaller but still clearly elevated tier** - all five inner-London, all five reported with their gaps flagged as upper bounds on the true redistributive effect, not point estimates.

**What a NaN means, and what it doesn't.** 63 standalone unitary authorities - including Hartlepool, County Durham and Blackpool, the Northern LAs this analysis' headline rests on - have no real peer group in the precepting-group lookup (they don't share a county, metropolitan-county, or GLA tier with anyone) and get `single_pot_bias_risk = NaN`. This is deliberate and load-bearing: **the statistic cannot vindicate them. It can only fail to indict them.** Absence of evidence is not evidence of absence. What actually supports "the Northern headline figures are largely clean" is `shared_tier_share` alone - Hartlepool 15.7%, County Durham 14.9%, Blackpool 13.9% - a real, independently computed bound that is unaffected by this gap: however much an unmeasured police/fire peer group might diverge, the shared-tier component can be at most `shared_tier_share` of the whole bill, whatever its true rate. That is a narrower, more defensible claim than "checked and clean," and is the one this analysis makes.

**A genuine bug was found and fixed while running this check, worth recording since it silently zeroed real revenue.** CTSOP suppresses small band counts as blank cells (disclosure control), parsed as `NaN`. The original code used `count or 0` to default missing counts to zero - but `NaN` is truthy in Python, so `nan or 0` evaluates to `nan`, not `0`. That NaN propagated into the tax-base sum, and pandas' default skip-NaN group-sum then silently collapsed an all-NaN group to exactly `0.0` - City of London read as **£0 actual revenue for 2017-18 onward** before the fix, discovered only because it's one of the five outlier LAs above and its risk score looked wrong. Fixed (`engine.build._safe_count`) and pinned with a regression test. Scope checked directly: 13 (LA, year) cells across the whole headline period, 2 LAs total (City of London and Tamworth) - small in aggregate, but not a coincidence that finding it mattered here specifically.

## Variant 3: did rate-setting already fix this?

This is the strongest form of the objection this whole repository exists to test. IFS (2020) themselves note it: because central government's funding allocation still references relative 1991 property values, a council in a high-appreciation area sits on a growing implicit tax base — it could, in principle, set a *lower* Band D rate and still raise the same revenue, quietly offsetting the freeze's redistributive effect without anyone touching council tax bands. If that's actually happening, the gap Variants 1 and 2 measure overstates the real-world effect. **This variant tests it directly, on real data, rather than leaving it as a caveat — and the answer strengthens the headline.**

**Panel: 2015-16 to 2025-26, not the full study window.** The control this regression needs — Core Spending Power, MHCLG's settlement-funding measure — doesn't exist as a concept before 2015-16 (earlier years used Formula Grant, then early Settlement Funding Assessment; reconstructing a comparable series would need the National Archives Web Archive, a materially harder extraction not pursued here). This is a different kind of gap from the 2009-wave CTSOP problem: Variant 3 tests a **behavioural regularity** — did councils respond to appreciation by cutting rates — not a cumulative total. Eleven years of within-LA variation across 296 LAs is an adequate panel to answer that on its own; the assumption this makes explicit is that the relationship, if any, is stable back to 2009 — weaker than several assumptions already accepted elsewhere in this pipeline.

**Specification**, every choice made explicitly: year-on-year *log growth* in Band D rate (not the level — not comparable across LAs with different service scopes, and LA fixed effects would absorb a level difference anyway) regressed on `relative_hpi` (the exact same `hpi_factor_la / hpi_factor_national` quantity Variants 1/2 already use — not a second definition of "relative appreciation"), controlling for year-on-year change in Core Spending Power per dwelling, with LA and year fixed effects, standard errors clustered by LA.

### The pooled result

| | coefficient | 95% CI | p | n |
|---|---|---|---|---|
| **Pooled, all 296 LAs** | **−0.0243** | **[−0.0318, −0.0167]** | **<0.0001** | 2,875 |

Taken at face value, this is exactly the IFS objection showing up: LAs with higher relative appreciation set systematically lower rate growth, tightly estimated. **A pooled result this clean is exactly the kind of number that should trigger a robustness check, not a press release** — and it doesn't survive one.

### The decomposition

| | coefficient | 95% CI | p | n |
|---|---|---|---|---|
| **Non-London (263 LAs)** | **−0.0063** | **[−0.0143, +0.0017]** | **0.125** | 2,545 |
| London only (33 LAs) | −0.0020 | [−0.0157, +0.0117] | 0.778 | 330 |

Excluding London drops the coefficient by 74% and its significance entirely. London on its own doesn't reach significance either (33 clusters is thin for cluster-robust standard errors) — the decisive number isn't a regression p-value at all, it's a simpler one: **the within-LA correlation between an LA's own year-to-year relative appreciation and its own rate growth is −0.57 in London, and +0.08 in the rest of England.** Not a weaker version of the same relationship — effectively zero, wrong sign, outside London. Region-level *average* rate growth is nearly flat across all nine regions (North East 4.27%, London 4.20%) — this was never a simple cross-region gradient. What's different about London is that individual boroughs' year-to-year appreciation swings are about **2.7x larger** than elsewhere (within-LA std 0.107 vs 0.039), giving them outsized leverage on a pooled coefficient even though the general-England relationship is close to flat.

**This is confounding, not compensation, and we already know the mechanism.** The pattern producing London's −0.57 is almost certainly the same one Phase 4 found and investigated directly (see "Validating the engine against IFS (2020)", above, and `engine.build.compute_single_pot_bias_risk`): Westminster/Wandsworth-type boroughs hold their own-tier rate artificially low on decades of business-rate income and reserves — a local fiscal anomaly with nothing to do with the valuation freeze — which correlates with London's high, volatile relative appreciation without being *caused* by it. Two independent investigations, a hand investigation of individual Band D rates in Phase 4 and this regression's robustness check in Phase 6, arrived at the same mechanism from two different directions.

### The finding

**The IFS objection is a claim about a general behavioural regularity: that councils in high-appreciation areas set lower rates, so the funding settlement already offsets the freeze's redistributive effect. Tested directly, across 263 non-London LAs and eleven years, no such regularity exists.**

The non-London specification is the primary result, not the pooled one — and it is a **tight** null, not a noisy one. The CI, [−0.0143, +0.0017], is narrow enough to rule out an economically meaningful offset, which was the standard this analysis committed to before running anything (see the pre-registration discipline in Sensitivity, above). Applied to the Northern headline LAs at the ceiling of what the non-London data supports:

| LA | implied offset, at the ceiling of the non-London CI |
|---|---|
| Hartlepool | 1.7% of measured Variant 1 gap |
| County Durham | 1.7% |
| Blackpool | 2.3% |

Even at the most generous reading the data allows, rate-setting explains a small fraction of the North's measured overpayment — not because the caveat is being minimised, but because that is what a tight null looks like when you compute it honestly.

**The pooled/London-driven coefficient must never be applied to a non-London LA — the single most misleading number this analysis could produce, and the code refuses to compute it** (`regression.variant3.interpret_offset` raises rather than extrapolating London's −0.57 relationship onto an LA where the measured relationship is +0.08). Applied instead to where it *is* measured — Westminster and Kensington and Chelsea, already flagged with the single-pot exposure caveat in Phase 4 — the pooled ceiling implies a sizeable apparent offset (27% and 36% of their measured gaps respectively). Reported prominently, not buried: a reader who runs the pooled regression themselves should find this number, and its diagnosis, already here.

### What this does to the headline

It strengthens it. We lead with the North. The regression clears the North: no evidence of rate-setting compensation, a tight null, not a shrug. The one place the compensating relationship does appear is London — the pole already caveated for single-pot exposure in Phase 4, and already reported as an upper bound rather than a point estimate. Three independent lines of evidence — the single-pot exposure bound, the Westminster rate-anomaly investigation, and this regression's London/non-London decomposition — all land on the same place, and it is not the place the headline number comes from.

## Sensitivity: is the headline robust, and what claim are we entitled to make?

Thresholds for this sweep were written down and committed **before** any grid point beyond the base case was computed — [SENSITIVITY_PREREGISTRATION.md](../SENSITIVITY_PREREGISTRATION.md) — because the purpose of a sensitivity sweep is to show the headline is robust and the base case conservative, which is exactly the answer everyone wants, and exactly the condition under which nobody scrutinises the result. Full results against those thresholds: [SENSITIVITY_RESULTS.md](../SENSITIVITY_RESULTS.md). Base case: North East +£227.63/dwelling/year, London −£394.92/dwelling/year (Variant 1, dwelling-year-weighted, 2009-10 to 2025-26).

**Headline: nothing crossed a pre-registered failure threshold.** No sign flip anywhere across the 12-cell midpoint grid, region-level HPI, or 5-/10-year revaluation cycles. The worst single movement anywhere in the whole sweep (10-year revaluation) still leaves the North East figure at 84% of its base-case size. That is the least interesting part of this section, and it is not where this section spends most of its words.

### The midpoint grid: the pre-registered direction was wrong, and it changes what claim we can make

The pre-registered prediction was that raising `BAND_A_RATIO` should *shrink* the North East gap and raising `BAND_H_RATIO` should *grow* it — reasoning that treated the liability curve (`engine.build._compressed_multiplier`) as fixed and asked only how a changed midpoint moves the value fed into it. **Both directions measured backwards.** North East is monotonically *increasing* in `BAND_A_RATIO` (£200.94 → £227.63 → £286.46 as it goes 0.6 → 0.75 → 0.9) and *decreasing* in `BAND_H_RATIO` (a much smaller effect: £228.48 → £226.44 across its full range). Monotonicity itself — the actual structural, pre-registered threshold — held exactly; only the direction of the a priori reasoning was wrong.

**Why:** the curve isn't fixed. `_multiplier_control_points` rebuilds it from these same two ratios every time, because it's defined to pass through each band's real statutory multiplier *at that band's own midpoint*. Raising `BAND_A_RATIO` moves both the assumed value *and* the curve's own anchor point simultaneously — a LA whose relative HPI trajectory sits close to the national average lands close to that anchor almost regardless of the ratio, so the ratio's effect runs mostly through LAs whose relative HPI diverges from the national average (which, by the thesis this repository tests, is exactly what Northern LAs do) via the curve's local slope, not a simple "more assumed value → bigger tax base" level effect. Neither of us saw this before running the sweep. That the pre-registration made it visible rather than letting it hide inside a plausible-looking result is the actual point of pre-registering.

**The consequence is a change in what claim this analysis is entitled to make, not a comment fix.** Since the Price Paid calibration section (above), the framing has been "our midpoint choices are conservative, so the measured gap is a floor" — built entirely on the (wrong) predicted direction. With the measured direction, is 0.75/1.5 still a lower bound? Answered jointly, not axis-by-axis, using the two *trusted* Price Paid calibration corners (thousands of sales each — Easington/Blackpool for Band A, Westminster/Kensington and Chelsea for Band H), run through the actual engine, not interpolated:

| Parameters | North East £/dwelling/year |
|---|---|
| `band_a_ratio=0.64, band_h_ratio=2.06` (Easington + Kensington and Chelsea — the corner that should minimise the gap under the measured direction) | **£205.79** |
| `band_a_ratio=0.75, band_h_ratio=1.50` (**base case**) | **£227.63** |
| `band_a_ratio=0.77, band_h_ratio=1.78` (Blackpool + Westminster — the corner that should maximise it) | **£232.30** |

Base sits *inside* this empirically-anchored range, but at 82% of the way from its floor (£205.79) to its ceiling (£232.30) — near the top, not the bottom. **We lose the floor framing.** The Price Paid calibration itself is unaffected — it is empirical evidence about where the true tails sit, independent of how the model responds to the parameter, and it still constrains the plausible range to roughly £206-232. What it no longer supports is the claim that our specific choice (0.75, 1.5) sits at the conservative end of that range. **The honest claim is a central estimate (£227.63/dwelling/year) with an empirically-anchored range of roughly £206-232, not "at least £206."** `config.py` and DATA.md have been corrected to say this plainly rather than carry the old claim forward as a stale comment.

### Collection factor: exact algebraic identity, functioning as a bug detector

`COLLECTION_FACTOR` is a uniform scalar applied to every LA's actual revenue, so the gap should scale *exactly* linearly with it — not an approximation, an algebraic identity (`gap[i] = C × (Z[i] − Y[i])` for a `C`-independent `Z, Y`). Measured: `gap(0.78)/gap(0.83) = 0.939759`, `gap(0.88)/gap(0.83) = 1.060241`, both regions, exact to 6 decimal places. This axis was pre-registered as a check for whether `COLLECTION_FACTOR` leaks into the pipeline non-uniformly somewhere — it detected no bug, which is the correct and intended non-finding here, not a substantive robustness result about the real world.

### HPI geography: a real but weaker result than predicted — reported as such, not inflated

The pre-registered prediction was that region-level HPI would *meaningfully dampen* the London figure specifically, on the reasoning that Westminster/Kensington and Chelsea plausibly diverge a lot from the London regional average. Measured: North East +0.5%, London +0.9% (more negative) — both close to a null result, nowhere near the pre-registered 50% threshold. **This is not a confirmation of the prediction, and should not be described as one.** A movement this small doesn't demonstrate that region-level HPI dampens London's heterogeneity, because it barely moved anything. What it does show, correctly stated: the finding is **not an artefact of using LA-level HPI granularity** — swapping to a coarser regional series barely changes the answer. That is a real, useful robustness result. It is a weaker claim than "region-level HPI dampens London as predicted," and the weaker claim is the one we make.

### Revaluation frequency: matched the prediction exactly

Periodic revaluation should lag continuous tracking and dampen (not reverse) the effect: continuous £227.63 → 5-year £211.32 (−7.2%) → 10-year £190.60 (−16.3%), monotonic, sign intact throughout, comfortably inside the pre-registered 40%-drop ceiling at 10 years. This is the one axis where the a priori mechanical reasoning and the measured result agree — reported the same way as the other three, not given extra weight for having been predicted correctly.
